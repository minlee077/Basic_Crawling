```python
from urllib.request import urlopen
import bs4

page_num = 1 # 페이지 넘버
```

```python
####################################################################################################################
#                            1페이지부터 마지막 페이지 까지 반복하기                                                   #
####################################################################################################################

main_url = "https://terms.naver.com/list.nhn?cid=51007&categoryId=51007" + "&page=" +str(page_num)
main_html = urlopen(main_url)

bs_obj = bs4.BeautifulSoup(main_html, "html.parser")           # 메인 페이지의 bs오브젝트 만들기
content_list=bs_obj.find("ul",{"class" : "content_list"})      # 메인페이지 안에 있는 질병들

subjects = content_list.findAll("div", {"class" : "subject"})
```

```python
####################################################################################################################
#                            페이지 안에서 각 질병에 대한 페이지로 들어가기                                             #
####################################################################################################################
pos=0
for subject in subjects:
    index = 0
    a_tag = subject.find("a")
    sub_url = "https://terms.naver.com" + a_tag.get('href')
    sub_html = urlopen(sub_url)                             # 각 질병 페이지의 URL


    bs_obj2 = bs4.BeautifulSoup(sub_html, "html.parser")    # 질병 페이지의 bs오브젝트 만들기

    disease_name = bs_obj2.find("title").text               # 질병의 이름
    p_tags = bs_obj2.findAll("p", {"class": "txt"})         # 질병에 대한 정보 (정의, 증상 등)
    h3s = bs_obj2.findAll("h3", {"class": "stress"})        # 증상, 원인, 하위질병 등

    JUNG="GARBAGEDATA"
    JENG="GARBAGEDATA"
    REAS="GARBAGEDATA"
    DIAG="GARBAGEDATA"
    TEST="GARBAGEDATA"
    CURE="GARBAGEDATA"
    PREV="GARBAGEDATA"
    
    for h3 in h3s:# 각징별의 정보 가져오기
        if(h3.text == '정의'):
            JUNG = p_tags[index].text
        elif (h3.text == '증상'):
            JENG = p_tags[index].text
        elif(h3.text == '원인'):
            REAS = p_tags[index].text
        elif (h3.text == '진단'):
            DIAG = p_tags[index].text
        elif (h3.text == '검사'):
            TEST = p_tags[index].text
        elif (h3.text == '치료'):
            CURE = p_tags[index].text
        elif (h3.text == '예방방법'):
            PREV = p_tags[index].text
        index += 1

        
    box = {}
    box ['1']=JUNG
    box ['2']=JENG
    box ['3']=REAS
    box ['4']=DIAG
    box ['5']=TEST
    box ['6']=CURE
    box ['7']=PREV
    print(pos,disease_name)
    print("정의")
    print(box['1'][0:20])
    print("증상")
    print(box['2'][0:20])
    print("원인")
    print(box['3'][0:20])
    print("진단")
    print(box['4'][0:20])
    print("검사")
    print(box['5'][0:20])
    print("치료")
    print(box['6'][0:20])
    print("예방")
    print(box['7'][0:20])
    print("----------------------")
    pos+=1

page_num += 1
```

# Problem

위 방식에서는 하나의 h3태그에 p태그가 없거나 p태그가 여러개 있을때 인덱스가 꼬이는 문제가 발생한다.

(ex) https://terms.naver.com/entry.nhn?docId=926877&cid=51007&categoryId=51007 치료에 대한 p txt태그가 없고 사진만 있는 경우)

따라서, findall이 아닌 다른 방식이 필요하다.


공식 Docs에 따르면,태그의 트리구조로 파싱이 가능함을 확인할 수 있다.

https://www.crummy.com/software/BeautifulSoup/bs4/doc/#navigating-the-tree


https://www.crummy.com/software/BeautifulSoup/bs4/doc/#next-sibling-and-previous-sibling

현재 사용할 문서에서 크롤링하려는 대상 특성상 트리에서 연속적인 위치의 태그를 확인하면 쉽게 해결할 수 있다.

따라서, bs에서 제공하는 next-sibling기능을 활용하면 해결 할 수 있다.
